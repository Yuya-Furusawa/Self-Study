```python
import numpy as np
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, download_plotlyjs
init_notebook_mode(connected=True)
```

# Cost Function

パラメータ$b$によるコスト関数の変化

\begin{equation}
    C(q_0, q_1) = b \log (e^{q_0/b} + e^{q_1/b})
\end{equation}

```python
def LMSR(q0, q1, b):
    return b * np.log(np.power(np.e, (q0 / b)) + np.power(np.e, (q1 / b)))
```

```python
x, y = np.meshgrid(np.linspace(0, 200, 32), np.linspace(0, 200, 32))
```

```python
b = 1
data = [
    go.Surface(
        z=LMSR(x,y,b)
    )
]
layout = go.Layout(
    title='LMSR : Cost Function',
    autosize=False,
    width=500,
    height=500,
    margin=dict(
        l=65,
        r=50,
        b=65,
        t=90
    )
)
fig = go.Figure(data=data, layout=layout)
URL = plotly.offline.iplot(fig, filename='Costfunction-3d-surface.html')
```
