# Cost Function

パラメータ$b$によるコスト関数の変化

\begin{equation}
    C(q_0, q_1) = b \log (e^{q_0/b} + e^{q_1/b})
\end{equation}

```python
def LMSR(q0, q1, b):
    return b * np.log(np.power(np.e, (q0 / b)) + np.power(np.e, (q1 / b)))
```

```python
x, y = np.meshgrid(np.linspace(0, 200, 32), np.linspace(0, 200, 32))
```

```python
b = 1
data = [
    go.Surface(
        z=LMSR(x,y,b)
    )
]
layout = go.Layout(
    title='LMSR : Cost Function',
    autosize=False,
    width=500,
    height=500,
    margin=dict(
        l=65,
        r=50,
        b=65,
        t=90
    )
)
fig = go.Figure(data=data, layout=layout)
URL = plotly.offline.iplot(fig, filename='Costfunction-3d-surface.html')
```

# Amount to pay

実際の支払額をみてみる

\begin{equation}
C(q_0 + x, q_1, b) - C(q_0, q_1, b)
\end{equation}

```python
# Buy x 10 units
buy_amount_x = 10
```

# Price

価格はコスト関数の微分値

\begin{equation}
p_i (q_i, q_j) = \frac{\exp(\frac{q_i}{b})}{\exp(\frac{q_i}{b}) + \exp(\frac{q_j}{b})}
\end{equation}

```python
# 0's price function
def PriceLMSR(x,y,b):
    return np.exp(x/b)/(np.exp(x/b) + np.exp(y/b))
```

```python
b = 1
data = [
    go.Surface(
        z=PriceLMSR(x,y,b)
    )
]
layout = go.Layout(
    title='LMSR : Price of X',
    autosize=False,
    width=500,
    height=500,
    margin=dict(
        l=65,
        r=50,
        b=65,
        t=90
    )
)
fig = go.Figure(data=data, layout=layout)
URL = plotly.offline.iplot(fig, filename='Costfunction-3d-surface.html')
```

## 価格の変化

```python
import matplotlib.pyplot as plt
```

```python
x_grid = np.arange(0, 200, 0.5)
params = [1, 10, 50, 100, 1000]
for p in params:
    price_vals = np.empty_like(x_grid)
    for i, x in enumerate(x_grid):
        price_vals[i] = PriceLMSR(x, 100, p)
    plt.plot(x_grid, price_vals, label="b = {}".format(p))
plt.grid()
plt.legend()
plt.show()
```
